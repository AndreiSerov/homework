```python
def violin_plot(data=pd.read_csv('diamonds.csv')):
    sns.set(
        rc={'figure.figsize': (18, 6)}
    )
    sns.violinplot(
        x='cut',
        y='price',
        data=data
    )
    sns.despine()
    plt.show()
```

```python
violin_plot(df)

```
