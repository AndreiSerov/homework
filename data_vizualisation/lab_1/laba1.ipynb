```python
from collections import Counter

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('diamonds.csv')
df
```

```python
sns.scatterplot(data=df, x="price", y="depth", color='orange')
plt.title('Точечная диаграмма по признакам price и depth')
plt.show()
```

```python
f, ax = plt.subplots()
sns.violinplot(data=df)
sns.despine(offset=10, trim=True);
```

```python
def violin_plot(data=pd.read_csv('diamonds.csv')):
    sns.set(
        rc={'figure.figsize': (18, 6)}
    )
    sns.violinplot(
        x='cut',
        y='price',
        data=data
    )
    sns.despine()
    plt.show()
```

```python
violin_plot(df)

```

```python
# PART 3 var 17
quakes = pd.read_csv('quakes.csv')
quakes
```

```python
#Круговая диаграмма
quakes_mag = Counter(quakes['mag'].round())
labels = quakes_mag.keys()
sizes = quakes_mag.values()
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Распределение Магнитуды Рихтера:')
plt.show()
```

```python
sns.pairplot(quakes)
```
