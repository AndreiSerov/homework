```python
def box_plot(data=pandas.read_csv('mtcars.csv')):
    plt.figure(figsize=(13, 10), dpi=80)
    sns.boxplot(x='cyl', y='disp', data=data, hue='am')

    for i in range(len(data['cyl'].unique()) - 1):
        plt.vlines(i + .5, 10, 45, linestyles='solid', colors='gray', alpha=0.2)
    plt.legend(title='as.factor(am)')
    plt.show()
```

```python
def violin_plot(data=pandas.read_csv('diamonds.csv')):
    sns.set(
        rc={'figure.figsize': (18, 6)},
        style="white"
    )
    sns.violinplot(
        x='cut',
        y='price',
        hue='cut',
        data=data
    )
    sns.despine()
    plt.show()
```

```python
violin_plot()
```

```python
def dot_plot(data=pd.read_csv('diamonds.csv')):
    data[['price', 'depth']].plot(
        kind='scatter',
        x='price',
        y='depth',
        figsize=(12, 8)
    )
    plt.show()
```
