```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
```

```python
df = pd.read_csv('diamonds.csv')
df
```

```python
#Гистограмма
sns.histplot(df,x="depth")
plt.title("Гистограмма для признака depth")
plt.show()
```

```python
# Гистограмма. График 2
sns.histplot(df,x="carat")
plt.title("Гистограмма для признака carat")
plt.show()
```

```python
#Точечная диаграмма. График 6
sns.scatterplot(data=df, x="price", y="carat", color='orange')
plt.title('Точечная диаграмма по признакам price и carat')
plt.show()
```
